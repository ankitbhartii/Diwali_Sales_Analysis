### As we can see here there are Null values at the end in row 13, 14, and we have to remove these Null values from dataset 


```python
#drop unrealted/blank columns
df.drop(['Status','unnamed1'], axis=1, inplace=True)
```

```python
df.info()
```

# Now we will here find out Null values

```python
#check for Null Values
pd.isnull(df).sum()
```

### there are 12 null values in Amount columns and now we will drop these null values from dataset


```python
df.shape
```

```python
#drop null values
df.dropna(inplace=True)
```

```python
df.shape
```

# Change Data Type

```python
#change Data Type
df['Amount'] = df['Amount'].astype('int')
```

```python
#review the data type has change or not after modfication
df['Amount'].dtype
```

# Exploratory Data Analysis


### Gender

```python
df.columns
```

```python
ax = sns.countplot(x = 'Gender',data = df, hue = 'Gender')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
grouped_df = df.groupby(['Gender'], as_index=False)['Amount'].sum()
sorted_df = grouped_df.sort_values(by='Amount', ascending=False)
sns.barplot(x= 'Gender', y = 'Amount',data = grouped_df, hue = 'Gender')
```

_From above graphs we can see that most of the buyers are females and Even the purchasing power of females are greater than men_

 # Age

```python
df. columns
```

```python
ax = sns.countplot(data = df, x = 'Age Group', hue = 'Gender')

for bars in ax.containers:
    ax.bar_label(bars)
    
```

```python
#Total amount Vs Age Group
grouped_df = df.groupby(['Age Group'], as_index=False)['Amount'].sum()
sorted_df = grouped_df.sort_values(by='Amount', ascending=False)
sns.barplot(x= 'Age Group', y = 'Amount',data = grouped_df)
```

_From above Graph most of the buyers belong to the Age Group betwwen 26-35 females_

# State

```python
df.columns
```

```python
#total numbers of amount/slaes from the top 10 states
grouped_df = df.groupby(['State'], as_index=False)['Orders'].sum()
sorted_df = grouped_df.sort_values(by='Orders', ascending=False).head(10)
sns.set_theme()
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(x= 'State', y = 'Orders',data = grouped_df, hue = 'State')

```

```python
#total numbers of orders from the top 10 states
grouped_df = df.groupby(['State'], as_index=False)['Amount'].sum()
sorted_df = grouped_df.sort_values(by='Amount', ascending=False).head(10)
sns.set_theme()
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(x= 'State', y = 'Amount',data = grouped_df, hue = 'State')

```

*From above Graphs we can see that  most of the orders are from Uttar Pradesh, Maharashtra and Karnataka respectively but total sales/amount is from up, Karnataka and then Maharashtra*  

# Marital status 

```python
ax = sns.countplot(data = df, x = 'Marital_Status')

sns.set(rc={'figure.figsize':(6,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
grouped_df = df.groupby(['Marital_Status','Gender'], as_index=False)['Amount'].sum()
sorted_df = grouped_df.sort_values(by='Amount', ascending=False)
sns.set_theme()
sns.set(rc={'figure.figsize':(6,5)})
sns.barplot(x= 'Marital_Status', y = 'Amount',data = grouped_df, hue= 'Gender')
```

*From above Graphs most of the Buyers are Married (Women) and they have High Purchasing Power*

# Occupation


```python
ax = sns.countplot(data = df, x = 'Occupation', hue= 'Occupation')

sns.set(rc={'figure.figsize':(20,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
grouped_df = df.groupby(['Occupation',], as_index=False)['Amount'].sum()
sorted_df = grouped_df.sort_values(by='Amount', ascending=False)
sns.set_theme()
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(x= 'Occupation', y = 'Amount',data = grouped_df, hue = 'Occupation')
```

*from above Graph as we can see that most of the buyers are working in IT, Aviation and  HealthCare sector*


# Product Category

```python
ax = sns.countplot(data = df, x = 'Product_Category', hue= 'Product_Category')

sns.set(rc={'figure.figsize':(30,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
grouped_df = df.groupby(['Product_Category',], as_index=False)['Amount'].sum()
sorted_df = grouped_df.sort_values(by='Amount', ascending=False)
sns.set_theme()
sns.set(rc={'figure.figsize':(25,5)})
sns.barplot(x= 'Product_Category', y = 'Amount',data = grouped_df, hue = 'Product_Category')
```

*from above Graphs we can see that most of the sold Products are from Food, Footwear,and Electronics Category*


```python
grouped_df = df.groupby(['Product_ID',], as_index=False)['Orders'].sum()
sorted_df = grouped_df.sort_values(by='Orders', ascending=False).head(10)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(x= 'Product_ID', y = 'Orders',data = grouped_df )

```
